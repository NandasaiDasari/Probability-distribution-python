```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
n=8
p=.5
k=np.arange(0,n+1)
binomial=stats.binom.pmf(k,n,p)         #pmf=probability mask function.
binomial                                #syntax=stats.binom.pmf(array of success values,number of trails,probability of success)
```

```python
plt.plot(k,binomial)                    
plt.title("Binomial distribution")
plt.xlabel("No of successes")
plt.ylabel("Probablity")
plt.show()
```

At average 3 persons visit bank in 1 minute at busy hours.
A)Probability for exactly four persons visiting in one minute.
B)Probability for more than 3 persons visiting in a minute.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
rate=3
k=np.arange(0,10)                       #since mean =3 let us take an array of size 10.(we can take any number)
poisson=stats.poisson.pmf(k,rate)       #pmf=probability mask function
poisson                                 #syntax=stats.poisson.pmf(array of possible values,mean)
```

```python
plt.plot(k,poisson)
plt.title("Poisson distribution")
plt.xlabel("No of people arriving")
plt.ylabel("Probablity")
plt.show()
```

```python
poisson[4]                     #probabilty of exactly 4 persons visiting in a minute.
```

```python
1-(poisson[0]+poisson[1]+poisson[2]+poisson[3])     #probability of more than 3 persons visiting in a minute.
```

```python
mean=0.295
std=0.025
x = np.linspace(mean - 3*std, mean + 3*std, 100)
stats.norm.pdf(x, mean, std)
```

```python
plt.plot(stats.norm.pdf(x, mean, std),x)
plt.title("Normal distribution")
```

```python
x=0.280                 # one way is use z-score.
z=(x-mean)/std          #since we have to find <0.280 we used x-mean.
z
```

```python
stats.norm.cdf(z)*100   #percentage cdf=cumulative distributive function,*100 for percentage.
                        #probability of weight less than 0.280kg
```

```python
x=0.350
z=(mean-x)/std          #since we have to find >0.350 we used mean-x.
z
```

```python
stats.norm.cdf(z)*100        #probability of weight>0.350kg
```

```python
stats.norm.cdf(0.280,loc=mean,scale=std)*100    # it gives probability for <x values with out using z-score.
```

```python
(1-stats.norm.cdf(0.350,loc=mean,scale=std))*100      # to find for >x use 1-(<x)
```

```python
(stats.norm.cdf(0.340,loc=mean,scale=std)-stats.norm.cdf(0.260,loc=mean,scale=std))*100     # for between two values.
#probability of weight between 0.260 to 0.340kg.
```
